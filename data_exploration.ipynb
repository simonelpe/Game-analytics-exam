```python
from pathlib import Path

import pandas as pd
from IPython.display import display

df = pd.read_csv("vgsales_clean.csv")
df
```

```python
df['User_Score'] = pd.to_numeric(df['User_Score'], errors='coerce')
df
```

```python
# convertire 'Year_of_Release' in data %Y
df['Year_of_Release'] = pd.to_datetime(df['Year_of_Release'], format='%Y', errors='coerce')
df
```

```python
# controlla quanti null per ogni singola colonna
df.isnull().sum()
```

```python
# rimuoviamo righe senza anno per i grafici temporali
df = df.dropna(subset=['Year_of_Release'])
df
```

```python
# Rimuoviamo righe gioco senza nome
df = df.dropna(subset=['Name'])
df
```

```python
# Sostituiamo i nan in publisher con 'Unknown'
df['Publisher'] = df['Publisher'].fillna('Unknown')
df
```

```python
# esporto in data.csv
df.to_csv("data.csv", index=False)
```
