```python
df['User_Score'] = pd.to_numeric(df['User_Score'], errors='coerce')
df
```

```python
# convertire 'Year_of_Release' in int
df['Year_of_Release'] = pd.to_numeric(df['Year_of_Release'], errors='coerce').astype('Int64')
df
```

```python
# controlla quanti null per ogni singola colonna
df.isnull().sum()
```

```python
# rimuoviamo righe senza anno per i grafici temporali
df = df.dropna(subset=['Year_of_Release'])
df
```

```python
# Rimuoviamo righe gioco senza nome
df = df.dropna(subset=['Name'])
df
```

```python
# Sostituiamo i nan in publisher con 'Unknown'
df['Publisher'] = df['Publisher'].fillna('Unknown')
df
```

```python
df.isnull().sum()
```

```python
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import ExtraTreesRegressor
```

```python
# 1. Pulizia User_Score
df['User_Score'] = pd.to_numeric(df['User_Score'], errors='coerce')

# --- MODIFICA RICHIESTA: RIEMPIRE DEVELOPER CON PUBLISHER ---
# Dove Developer è NaN, copiamo il valore di Publisher
print(f"Developer mancanti prima: {df['Developer'].isna().sum()}")
df['Developer'] = df['Developer'].fillna(df['Publisher'])
print(f"Developer mancanti dopo il fix 'Publisher': {df['Developer'].isna().sum()}")
# ------------------------------------------------------------

# Creiamo la copia di lavoro
df_impute = df.copy()
if 'Name' in df_impute.columns:
    df_impute = df_impute.drop(columns=['Name'])

# 2. Mappatura MANUALE Rating (Logica per età)
rating_map = {
    'EC': 0, 'E': 1, 'K-A': 1, 'E10+': 2, 'T': 3, 
    'M': 4, 'AO': 5, 'RP': 6
}
inv_rating_map = {v: k for k, v in rating_map.items()}
df_impute['Rating'] = df_impute['Rating'].map(rating_map)

# 3. Encoding Automatico
encoders = {}
cols_auto_encode = ['Platform', 'Genre', 'Publisher', 'Developer'] 

for col in cols_auto_encode:
    le = LabelEncoder()
    # Fit su tutti i valori presenti (ora Developer ha molti più dati grazie al Publisher)
    mask = df_impute[col].notnull()
    # Convertiamo in stringa per uniformità
    df_impute.loc[mask, col] = le.fit_transform(df_impute.loc[mask, col].astype(str))
    encoders[col] = le

print("Dati pronti. Developer integrato con Publisher e Rating mappato.")
```

```python
# Usiamo ExtraTreesRegressor (Foreste decisionali)
# Questo capirà che Developer "Nintendo" -> Rating "E" (1)
model = ExtraTreesRegressor(n_estimators=50, n_jobs=-1, random_state=42)

imputer = IterativeImputer(estimator=model, max_iter=10, random_state=42)

print("Inizio imputazione intelligente...")
df_filled_array = imputer.fit_transform(df_impute)

df_filled = pd.DataFrame(df_filled_array, columns=df_impute.columns, index=df_impute.index)
print("Fatto.")
```

```python
# 1. Arrotondamenti
df_filled['Critic_Count'] = df_filled['Critic_Count'].round().astype(int)
df_filled['User_Count'] = df_filled['User_Count'].round().astype(int)
df_filled['Critic_Score'] = df_filled['Critic_Score'].round().astype(int)
df_filled['User_Score'] = df_filled['User_Score'].round(1)

# 2. Decodifica Categorie (Developer incluso)
for col in cols_auto_encode:
    df_filled[col] = df_filled[col].round().astype(int)
    max_val = len(encoders[col].classes_) - 1
    df_filled[col] = df_filled[col].clip(0, max_val)
    df_filled[col] = encoders[col].inverse_transform(df_filled[col])

# 3. Decodifica Rating (Manuale)
df_filled['Rating'] = df_filled['Rating'].round().astype(int).clip(0, 6)
df_filled['Rating'] = df_filled['Rating'].map(inv_rating_map)

# 4. Unione con Name
if 'Name' in df.columns:
    df_final = df_filled.copy()
    df_final.insert(0, 'Name', df['Name'])
else:
    df_final = df_filled

print("Controllo finale su Super Mario Bros (dovrebbe avere Dev=Nintendo e Rating=E):")
display(df_final[df_final['Name'] == 'Super Mario Bros.'].head())
```

```python
print("Valori nulli rimanenti:")
print(df_final.isnull().sum())
```

```python
# esporto in data.csv
df_final.to_csv("data.csv", index=False)
```
